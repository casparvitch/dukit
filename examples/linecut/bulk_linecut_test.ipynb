```python
import matplotlib.pyplot as plt
import numpy as np
from dukit import BulkLinecutWidget
%matplotlib tk
```

```python
fig, axs = plt.subplots(ncols=3, figsize=(12, 6))

data = np.loadtxt("./linecut_test_image.txt")
data_mean = np.mean(data)
images = [data + 25*data_mean*np.random.rand(*data.shape) for _ in range(5)]
image_names = [str(i) for i, _ in enumerate(images)]

selector_image = images[-1] # pick image to put in gui for you to select from 
axs[0].imshow(selector_image)  # (data may be nans if you want empty selector)

selector = BulkLinecutWidget(*axs, images, image_names)
plt.show(block=True)

opath = None # "./linecut_output.json"
selector.disconnect(path=opath) # if path is None it prints output 
```
